You can also plot complex numbers. You can plot using either cartesian or polar coordinates. The example below shows how you can plot the difference of two complex number z, y (x) in cartesian coordinates

```python
z = complex(10,10)
y = complex(2,4)
x = z - y

numbers = [x,y,z]

for x in numbers:
    r, phi = cmath.polar(x)
    plt.polar([0,phi],[0,r], marker='o')
    
print(x)
```

The most importan thing to note above is that you can think of a complex number that create a point on a circle where phi is the angle and r is the radius of the circle. Later we will take this to three dimensions using vectors to create points a 3D Sphere (also called the Bloch Sphere.

1) For the given list of complex numbers can you add them all together? 

```python
complex_numbers = [complex(1,1), complex(2,1), complex(6,3), complex(-3,4), complex(6,-1)]
```

```python
# The simplest way

total = 0
for number in complex_numbers:
    total = total + number

total
# complex_numbers[0] + complex_numbers[1]
```

Another solution using fancy 3.8 list comprehension and the walrus operator

```python
total = 0
[total := total + x for x in complex_numbers]
total
```

```python
complex_numbers.append(total)
```
